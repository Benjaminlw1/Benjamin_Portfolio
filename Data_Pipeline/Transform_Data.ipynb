```python
import os
import pandas as pd
```

```python
cwd = os.getcwd() # dirname()
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'Data_Pipeline/data')
CACHE_DIR = os.path.join(BASE_DIR, 'Data_Pipeline/cache')
work_file = os.path.join(CACHE_DIR, 'ranking_movies_dataset.csv')
out_file = os.path.join(CACHE_DIR, 'ranking_movies_dataset_transform.csv')
```

```python
df = pd.read_csv(work_file)
```

```python
df.head(n=1000)
```

```python
df['Rank'] = -1
```

```python
df.head()
```

```python
df['Domestic %'] = df['%']
df['Foreign %'] = df['%.1']
df.head()
```

```python
df.drop(columns=["Unnamed: 0",'%','%.1'],inplace=True)
```

```python
prep_clean_cols = ['Worldwide','Domestic','Foreign']
def convert_str_to_int(currency_val):
    currency_val = currency_val.replace("$","").replace(",","")
    try:
        currency_val = int(currency_val)
    except:
        # turns "-" into int 0
        currency_val = 0
        
    return currency_val

def clean_col(row):
    """
    row = pandas series
    dataframe = collection of Pandas Series
    """
    #print(row)
    for col in prep_clean_cols:
        current_val = row[col]
        row[col] = convert_str_to_int(current_val)
    return row

df = df.apply(clean_col, axis=1)
```

```python
df.dtypes
```

```python
df.sort_values(by=['Worldwide'], inplace=True, ascending=False)
df.head()
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.head()
```

```python
df['Rank'] = df.index + 1
df.head()
```

```python
df['Domestic %'] = df['Domestic'] / df['Worldwide']
df['Foreign %'] = df['Foreign'] / df['Worldwide']
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.to_csv(out_file, index=False)
```
