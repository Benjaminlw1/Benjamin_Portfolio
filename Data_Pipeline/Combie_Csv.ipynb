```python
import os
import pandas as pd
```

```python
cwd = os.getcwd() # dirname()
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'Data_Pipeline/data')
CACHE_DIR = os.path.join(BASE_DIR, 'Data_Pipeline/cache')
os.makedirs(CACHE_DIR, exist_ok=True)
```

```python
my_dataframes = []
csv_files = [x for x in os.listdir(DATA_DIR) if x.endswith(".csv")]

for filename in csv_files:
    print(filename)
    year = filename.replace(".csv", "")
    csv_path = os.path.join(DATA_DIR, filename)
    this_df = pd.read_csv(csv_path)
    this_df['filename'] = filename
    this_df['year'] = year
    print(this_df.head(n=1))
    my_dataframes.append(this_df)
```

```python
my_dataframes
```

```python
combine_dataframe = pd.concat(my_dataframes)
```

```python
combine_dataframe.tail()
```

```python
dataset = os.path.join(CACHE_DIR, 'ranking_movies_dataset.csv')
combine_dataframe.to_csv(dataset)
```
